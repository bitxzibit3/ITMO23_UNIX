# Лабораторная работа 3. Халимов Тимур, К3332.

## Задание 1. Создать программу CALC с интерфейсом CLI
Создать программу под названием CALC, которая будет принимать на вход 3
аргумента (2 операнда и 1 оператор). Оператор может быть: +, -, ^. Реализация
операторов только с использованием функций, условий, циклов, +, - и *.

В папке `src` создадим файл `calc.cpp`. Сборку осуществим при помощи утилиты `g++`:
```bash
g++ -o build/calc src/calc.cpp
```
